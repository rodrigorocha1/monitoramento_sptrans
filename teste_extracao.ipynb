```python
from src.api.api import API
import os
import pandas as pd
from time import sleep
from random import randint
import requests
```

```python
def gravar_dados_linha(controle_indice):
    lista_temp=  []
    caminho = os.getcwd() + '/data/datalake/bronze/arquivos_gtfs/routes.txt'

    base_linha = pd.read_csv(caminho)
    for indice, valor in base_linha.iterrows():
        print(f'indice {indice + 1}')
    
        if indice + 1 >= controle_indice:
            tempo = randint(1, 100)
            print(f'Ã­ndice: {indice + 1} aguarde {tempo} segundos')
            try:
                sleep(tempo)
                dados_linha = API.ober_dados_linha(valor['route_id'])
                dados_dict = {
                    'LINHA': valor['route_id'],
                    'CODIGO_AREA': dados_linha[1],
                    'CONSOCIO': dados_linha[2],
                    'EMPRESA': dados_linha[3]
                }
                print(valor['route_id'])
                if indice + 1 == 600:
                    break
                lista_temp.append(dados_dict)
                df = pd.DataFrame(lista_temp)
                df.to_csv(os.getcwd() + f'/data/datalake/bronze/relacao_empresas_{controle_indice}.csv')
            except:
                print('falha', valor['route_id'])

            
```

```python
gravar_dados_linha(500)
```
